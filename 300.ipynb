```python
a = 'hihellohohahuhahhhh'
count = 0
result = ''

for c in a :
    if c == 'h' :
        result += c.upper()
        count += 1
    else :
        result += c

    if count == 3 :
        break

print(result)    
```

```python
words = ["apple", "banana", "cherry", "date", "elderberry"]

a = [st for st in words if len(st) >= 5]

print(a)
```
