```python
y = [10, 20, 30]

try :
    index, x = map(int, input('인덱스, 나눌 숫자').split())
    print(y[index] / x)
except ZeroDivisionError as e :
    print('can\'t divide 0', e)
except IndexError as e:
    print('wrong index')
else :
    print('success')
finally :
    print('end')
```

```python
from abc import abstractmethod
class Bird :

    @abstractmethod
    def fly(self) :
        pass
    
class Eagle(Bird) :

    def display(self) :
        print('display')

    def fly(self) :
        print('fast')

    
eagle = Eagle()
eagle.fly()
```

```python
from abc import ABC, abstractmethod
class Bird(ABC) : # 추상 클래스 만들기

    @abstractmethod
    def fly(self) :
        pass
    
class Eagle(Bird) :

    def display(self) :
        print('display')

    # def fly(self) :
    #     print('fast')

    
eagle = Eagle()
eagle.fly()
```

```python
class UserClass(Exception) :
    def __str__(self):
        return 'can\'t use'
```

```python
try : 
    x = int(input('number : '))
    if x <= 0 :
        raise UserClass
    print(x)
except Exception as e :
    print(e)
except UserClass as e:
    print(e)
except ValueError as e:
    print(e)
```
