```python
import json
from datetime import datetime

data = {
    'c_group' : 'ai 데이터',
    'a_id' : 1004,
    'd_data' : [1, 2, 3],
    'b_date' : datetime.now()
}

print(datetime.now())

def json_handler(obj) :
    if isinstance(obj, datetime) :
        return obj.strftime('%Y-%m-%d : %H-%M-%S')

result = json.dumps(
    data,
    indent = 4,
    ensure_ascii = False,
    sort_keys = True,
    default = json_handler
)

print(result)
```

```python
import json

header = ['학번', '이름', '전공']
students = [
    ['202601', '홍', '컴'],
    ['202602', '김', '데'],
    ['202603', '유', '인']
]

data = [dict(zip(header, student)) for student in students]

# print(data)

j_data = json.dumps(
    data,
    indent=4,
    ensure_ascii=False,
    sort_keys=False
    )

print(j_data)
```

```python
with open('../source/students.json', 'w', encoding='utf-8') as f :
    json.dump(data, f, ensure_ascii = False, sort_keys=True)
```

```python
with open('../source/students.json', 'r', encoding='utf-8') as f :
    load_d = json.load(f)

print(load_d)
print(type(load_d))
```

```python
import csv

with open('../source/records.csv', 'w', encoding='utf-8-sig', newline='') as f :
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(students)

with open('../source/records.csv', 'r', encoding='utf-8-sig') as f :
    reader = csv.reader(f)
    # print(reader)
    for row in reader :
        # print(row)
        print(f'학번 : {row[0]} | 이름 : {row[1]} | 전공 : {row[2]}')
```
