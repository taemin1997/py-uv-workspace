```python
class Book :

    def __init__(self, title = None, auth = None, page = None) :
        self.__title = title
        self.__auth = auth
        self.__page = page
    
    def getTitle(self) :
        return self.__title
    
    def getAuth(self) :
        return self.__auth
    
    def getPage(self) :
        return self.__page
    
    def setTitle(self, title) :
        self.__title = title

    def setAuth(self, auth) :
        self.__auth = auth

    def setPage(self, page) :
        self.__page = page

    def toString(self) :
        return f'the author of {self.getTitle()} is {self.getAuth()} and {self.getPage()} pages'
```

```python
class BookRun : 

    def process(self) :
        bk1 = Book()
        bk1.setTitle('python')
        bk1.setAuth('kim')
        bk1.setPage(222)
        print(bk1.toString())

if __name__ == '__main__' :
    br = BookRun()
    br.process()
```

```python
class Dog :

    count = 0

    def __init__(self, name) :
        self.name = name
        Dog.count = Dog.count + 1

    @classmethod
    def get_count(cls) :
        return cls.count
    
    @staticmethod
    def is_valid_name(name) :
        return isinstance(name, str) and len(name) > 0
        
    
```

```python
dog1 = Dog('d1')
print(dog1.get_count())
dog2 = Dog('d2')
print(Dog.get_count())
```

```python
class Parent:
  x = 'parent x'  #클래스 변수
  y = 'parent y'  #클래스 변수


  @classmethod
  def output(cls):
    return cls.x


  def test(self):
     return 10


```

```python
class Child(Parent):
  x = 'child x' # 클래스 변수


  def display(self):
    print('display')
    print(Parent.output())
    print(Child.output())
    print(Parent.test(self))
    print(Child.test(self))




print(Parent.x)
print(Parent.y)


print(Child.x)
print(Child.y)
```

```python
class User:
  def __init__(self, name, email):
    self.name = name
    self.email = email


class AdminUser(User):
  def __init__(self, name, email, role):
    super().__init__(name, email)
    self.role = role
```

```python
admin = AdminUser('Alice', 'alice@gmail.com', 'admin')
print(admin.name)
print(admin.email)
print(admin.role)
```

```python
from abc import ABC, abstractmethod


class Shape(ABC):


    def __init__(self, width, height):
        self.width = width
        self.height = height


    @abstractmethod
    def area(self):
      pass    




class Rect(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)


    def area(self):
      return self.width * self.height  


   
class Triangle(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)


    def area(self):
      return self.width * self.height * 0.5
```
