문제 1: 클래스 메서드로 인스턴스 카운트 관리  
요구사항:
User 클래스를 만들고, 사용자가 생성될 때마다 클래스 변수 count를 1씩 증가시키세요.
get_user_count()라는 클래스 메서드로 현재 사용자의 수를 반환할 수 있어야 합니다.

```
class User:
    # 클래스 변수 count 정의

    def __init__(self, name):
        self.name = name
        # 여기에서 사용자 수 증가

    # classmethod get_user_count() 작성

# 사용 예시
u1 = User("Alice")
u2 = User("Bob")
print(User.get_user_count())  # 출력: 2

```

````python
class User:
    count = 0  # 클래스 변수

    def __init__(self, name):
        self.name = name
        User.count += 1  # 클래스 변수 증가

    @classmethod
    def get_user_count(cls):
        return cls.count  # cls를 통해 접근 가능

# 테스트
u1 = User("Alice")
u2 = User("Bob")
print(User.get_user_count())  # 출력: 2
````

    # staticmethod is_valid_email() 작성

문제 3: 클래스 메서드로 객체 초기화 조건 분기  
요구사항:
Circle 클래스는 반지름 radius를 받습니다.
**from_diameter(cls, diameter)**라는 클래스 메서드를 작성하여, 지름으로부터 객체를 생성할 수 있게 하세요.
```
class Circle:
    def __init__(self, radius):
        self.radius = radius

    # classmethod from_diameter 작성

# 사용 예시
c = Circle.from_diameter(10)
print(c.radius)  # 출력: 5.0

```


````python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    @classmethod
    def from_diameter(cls, diameter):
        return cls(diameter / 2)  # cls로 객체 생성

# 테스트
c = Circle.from_diameter(10)
print(c.radius)  # 출력: 5.0

````

문제 5: 정적 메서드를 활용한 포맷 유틸리티  
요구사항:
Formatter 클래스를 만들고, @staticmethod로 다음 기능을 구현하세요.

to_snake_case(text) : 문자열을 snake_case로 변환
(예: "Hello World" → "hello_world")
```
class Formatter:
    # staticmethod to_snake_case 작성

# 사용 예시
print(Formatter.to_snake_case("Hello World"))  # 출력: hello_world
print(Formatter.to_snake_case("Python Programming"))  # 출력: python_programming

```

````python
class Formatter:
    @staticmethod
    def to_snake_case(text):
        return text.lower().replace(" ", "_")

# 테스트
print(Formatter.to_snake_case("Hello World"))        # hello_world
print(Formatter.to_snake_case("Python Programming")) # python_programming
````

문제 3: super() 사용  
요구사항:
Employee 클래스를 만들고 생성자에서 name과 id를 초기화합니다.
Manager 클래스는 Employee를 상속받고, 생성자에서 department까지 추가로 초기화하도록 하세요.
super()를 이용해 부모 클래스의 생성자를 호출해야 합니다.

```
class Employee:
    def __init__(self, name, id):
        self.name = name
        self.id = id

class Manager(Employee):
    # 생성자에서 name, id, department 초기화
    # super() 사용

# 사용 예시
m = Manager("Alice", 101, "Sales")
print(m.name)        # 출력: Alice
print(m.department)  # 출력: Sales

```


````python
class Employee:
    def __init__(self, name, id):
        self.name = name
        self.id = id

class Manager(Employee):
    def __init__(self, name, id, department):
        super().__init__(name, id)  # 부모 클래스 생성자 호출
        self.department = department

# 테스트
m = Manager("Alice", 101, "Sales")
print(m.name)        # 출력: Alice
print(m.department)  # 출력: Sales
````

문제 1: 오버라이딩 + super()의 활용  
요구사항:
Shape 클래스에 area() 메서드를 만들고, Rectangle에서 오버라이딩하세요.
오버라이딩된 메서드 내에서 super().area()를 호출해 "Calculating area..."를 출력한 뒤 실제 넓이를 반환하도록 하세요.
```
class Shape:
    def area(self):
        print("Calculating area...")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # area 메서드 오버라이딩
    # super().area() 호출 포함

# 사용 예시
r = Rectangle(3, 4)
print(r.area())  # 출력: Calculating area... \n 12
```

````python
class Shape:
    def area(self):
        print("Calculating area...")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        super().area()  # 부모 메서드 호출
        return self.width * self.height

# 테스트
r = Rectangle(3, 4)
print(r.area())  # 출력: Calculating area... \n 12
````

 문제 4: 추상 클래스(ABC) 사용  
요구사항:
Vehicle은 추상 메서드 drive()를 가진 추상 클래스입니다.
Car 클래스는 이를 상속받고 drive()를 구현해야 합니다.
미구현 시 오류가 발생하는지도 확인해보세요.
```
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def drive(self):
        pass

class Car(Vehicle):
    # drive 메서드 구현

# 사용 예시
c = Car()
c.drive()  # 출력 예: Driving...
```

문제 5: isinstance()와 issubclass() 활용  
요구사항:
Animal, Cat, Dog 클래스를 정의하고 다음을 확인하세요:

Cat은 Animal의 서브클래스인가?

Cat() 인스턴스는 Animal의 인스턴스인가?

isinstance(Cat(), Dog) 결과는?
```
class Animal:
    pass

class Cat(Animal):
    pass

class Dog(Animal):
    pass

# 출력:
print(issubclass(Cat, Animal))       # True
print(isinstance(Cat(), Animal))     # True
print(isinstance(Cat(), Dog))        # ?
```

````python
class Animal:
    pass

class Cat(Animal):
    pass

class Dog(Animal):
    pass

print(issubclass(Cat, Animal))       # True
print(isinstance(Cat(), Animal))     # True
print(isinstance(Cat(), Dog))        # False
````
