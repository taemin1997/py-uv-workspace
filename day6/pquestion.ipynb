```python
class Stock :
    
    def __init__(self, name, code, PER, PBR, rate) :
        self.__name = name
        self.__code = code
        self.__PER = PER
        self.__PBR = PBR
        self.__rate = rate

    def getName(self) :
        return self.__name
    
    def getCode(self) :
        return self.__code
    
    def getPBR(self) :
        return self.__PBR

    def getPER(self) :
        return self.__PER

    def setPBR(self, PBR) :
        self.__PBR = PBR

    def setPER(self, PER) :
        self.__PER = PER

    def setCode(self, code) :
        self.__code = code

    def setName(self, name) :
        self.__name = name

    def getRate(self) :
        return self.__rate

    def setRate(self, rate) :
        self.__rate = rate
```

```python
sam = Stock('samsung', '005930', 15.79, 1.33, 2.83)

sam.setPER = 12.75
```
