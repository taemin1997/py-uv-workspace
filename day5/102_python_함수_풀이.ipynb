[문제2]   
문자열을 입력받아 각 문자들로 구성된 리스트로 반환하는 make_list 함수를 정의하세요.  
make_list("abcd")  

[출력결과]

['a', 'b', 'c', 'd']


```python
def make_list (sn) :
    my_list = []
    for data in sn :
        my_list.append(data)
    return my_list
```

```python
make_list("abcd")
```

[문제3]   
숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하세요.   

pickup_even([3, 4, 5, 6, 7, 8])

[출력결과]   
[4, 6, 8]

```python
def pickup_even(items):
    result = []
    for item in items:
        if item % 2 == 0:
            result.append(item)
    return result
```

```python
pickup_even([3, 4, 5, 6, 7, 8])
```

[문제4]  
세 개의 숫자를 입력받아 가장 큰수를 출력하는 print_max 함수를 정의하라. 단 if 문을 사용해서 수를 비교하세요.   
print_max(10, 2, 5)

[출력결과]  
10

```python
def print_max(a, b, c) :
    max_val = 0
    if a > max_val :
        max_val = a
    if b > max_val :
        max_val = b
    if c > max_val :
        max_val = c
    print(max_val)
```

```python
print_max(10, 2, 5)
```

[문제5]    
2 와 5의 배수가 아닌 갯수와 합계를 구하는 함수를 구현하세요.    

result(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)  

[출력결과]    
2 와 5의 배수가 아닌 갯수 : 7    
2 와 5의 배수가 아닌 합계 : 107   

==============================================  
result(5,2,7,9,10,8,6)     

[출력결과]    
2 와 5의 배수가 아닌 갯수 :  2  
2 와 5의 배수가 아닌 합계 :  16



```python
def result(*data):
    cnt = 0
    total = 0

    for i in data:
        if i % 2 != 0 and i % 5 != 0:
            cnt +=  1
            total += i

    print("2 와 5의 배수가 아닌 갯수 : ", cnt)
    print("2 와 5의 배수가 아닌 합계 : ", total)
```

```python
result(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)
```

```python
result(5,2,7,9,10,8,6)
```

문제6~13) 아래 함수를 호출했을때 정상적으로 실행이 되도록 호출하는 함수를 구현하세요.

```python
# 문제6)
def calc(a, b, /, c, *, d):
    return a + b + c + d
```

```python
print(calc(1, 2, 3, d=4))
```

```python
# 문제7)
def introduce(name, /, age, *, country):
    return f"{name}, {age} years old from {country}"
```

```python
introduce("Alice", 30, country="USA")
```

```python
# 문제8)
def f(x, *args, y, **kwargs):
    return x, args, y, kwargs
```

```python
f(1, 2, 3, y=4, z=5)
f(1, y=2)
```

```python
# 문제9)
def display(*args, **kwargs):
    print("ARGS:", args)
    print("KWARGS:", kwargs)
```

```python
display(1, 2, 3, name="Tom", age=30)
```

```python
# 문제10)
def config(*, debug=False, verbose=False):
    return f"Debug={debug}, Verbose={verbose}"
```

```python
config(debug=True, verbose=True)
```

```python
# 문제11)
def info(name, /, *, age, country):
    return f"{name}, {age}, {country}"
```

```python
#name: positional-only
# age, country: keyword-only
info("Tom", age=25, country="Canada")
```

```python
# 문제12)
def combine(a, /, b, *, c):
    return a + b + c
```

```python
combine(1, 2, c=3)
combine(1, b=2, c=3)
```
