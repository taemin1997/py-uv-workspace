```python
def proc(*a) :
    total = 0
    avg = 0

    for v in a :
        total += v
    
    avg = total / len(a)

    return total, avg
```

```python
total , avg = proc(1, 2, 3, 4)
print(f'total : {total}')
print(f'avg : {avg}')
```

```python
k = 1
def outer() : 
    m = 1
    def inner() : 
        global k
        k += 1
        nonlocal m
        m += 1

        return k, m
    return inner()
```

```python
print(outer())
```

```python
def fac(n) : 
    if n <= 1 :
        return 1

    return n * fac(n - 1)
```

```python
print(fac(5))
```

```python
from functools import lru_cache

@lru_cache(maxsize = None) 
def proc_fac(n) :
    if n <= 1 :
        return 1
    return n * proc_fac(n - 1)
```
