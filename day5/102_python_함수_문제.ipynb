[문제2]   
문자열을 입력받아 각 문자들로 구성된 리스트로 반환하는 make_list 함수를 정의하세요.  
make_list("abcd")  

[출력결과]

['a', 'b', 'c', 'd']


```python
def m_list(a) :
    li = []
    for ch in a :
        li.append(ch)
    return li

print(m_list('avafa'))
```

[문제3]   
숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하세요.   

pickup_even([3, 4, 5, 6, 7, 8])

[출력결과]   
[4, 6, 8]

```python
def pick(a) :
    return [n for n in a if n % 2 == 0]

print(pick([2, 3, 4, 5, 1, 6]))
```

max(10, 2, 5)

[출력결과]  
10

```python
def f_max(a, b, c) :
    if a >= b and a >= c :
        return a
    elif b >= a and b >= c :
        return b
    else :
        return c
```

[문제5]    
2 와 5의 배수가 아닌 갯수와 합계를 구하는 함수를 구현하세요.    

resultr(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)  

[출력결과]    
2 와 5의 배수가 아닌 갯수 : 7    
2 와 5의 배수가 아닌 합계 : 107   

==============================================  
result(5,2,7,9,10,8,6)     

[출력결과]    
2 와 5의 배수가 아닌 갯수 :  2
2 와 5의 배수가 아닌 합계 :  16



```python
def proc(a) :
    count = 0
    total = 0
    for n in a :
        if n % 5 != 0 and n % 2 != 0 :
            count += 1
            total += n
            
    print(total, count)    
b = [1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31]

proc(b)
```
