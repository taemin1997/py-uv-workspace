 - 코랩은 웹브라우저에서 텍스트와 프로그램을 작성할 수 있는 온라인 텍스트 에디터 이다.
 - 클라우드 기반의 주피터 노트북 개발환경이다.
 - 코랩은 텍스트를 입력할 수 있는 텍스트 셀과 코드를 입력할 수 있는 코드셀를 제공한다.

### [셀상태]
 - edit mode(입력 모드)
   - Edit Mode는 셀 안에 코드를 입력할 수 있다.
   - 셀 내부를 마우스 오른쪽하거나 Command Mode에서 Edit Mode 진입 단축키(Enter)를 통해서 Mode를 전환 할 수 있다.
    
 - command mode(명령형 모드)
   - Command mode는 셀 경계에서 단축키를 사용할  수 있다.
   - 셀 바깥의 여백에 대하여 마우스 우클릭하거나, Edit Mode에서 Command Mode 진입 단축키 (ESC )을 통해서 mode를 전환할 수 있다.

### [실행 방법]
 - Ctrl + Enter : 실행하고 커서가 현재 위치에 그대로 있다.
 - Alt + Enter : 실행하고 다음 셀을 생성해 준다.
 - Shift + Enter : 실행하고 이동할 셀이 있으면 이동만하고 이동할 셀이 없으면 새로 셀을 생성한다.

### [ 단축키]
 - Ctrl + M A  : 코드 셀 위에 삽입
 - Ctrl + M B  : 코드 셀 아래 삽입
 - Ctrl + M D  : 셀 삭제
 - Ctrl + M M  : 마크다운(텍스트)셀로 변경
 - Ctrl + M Y  : 코드셀로 변경
 - Ctrl + M Z  : 실행 취소
 - Ctrl + M H  : 단축키 모음을 보여줌

* MAC을 사용하는 경우 Ctrl 대신 Command사용

### [ 입력 및 편집 이동]
 - 입력 -> 편집 이동 : ESC
 - 편집 -> 입력 이동 : Enter
