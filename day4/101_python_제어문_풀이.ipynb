[문제10]
if~else문을 이용해서 년도에 해당하는 띠를 구하는 프로그램을 구현하시오.
단, 년도은 input()으로 입력받음.

12가지 띠
원숭이, 닭, 개, 돼지, 쥐, 소, 호랑이, 토끼, 용, 뱀, 말, 양

띠구하는 공식
year%12의 결과가
0이면 "원숭이", 1이면 "닭" 2이면 "개", 3이면 "돼지",
4이면 "쥐", 5이면 "소", 6이면 "호랑이", 7이면 "토끼",
8이면 "용", 9이면 "뱀", 10이면 "말", 11이면 "양"

출력결과
2012년도의 태생은 용띠입니다.

```python
# 연도 입력 받기
year = int(input("연도를 입력하세요: "))

# year % 12 값을 기반으로 띠 계산
remainder = year % 12

if remainder == 0:
    zodiac = "원숭이"
elif remainder == 1:
    zodiac = "닭"
elif remainder == 2:
    zodiac = "개"
elif remainder == 3:
    zodiac = "돼지"
elif remainder == 4:
    zodiac = "쥐"
elif remainder == 5:
    zodiac = "소"
elif remainder == 6:
    zodiac = "호랑이"
elif remainder == 7:
    zodiac = "토끼"
elif remainder == 8:
    zodiac = "용"
elif remainder == 9:
    zodiac = "뱀"
elif remainder == 10:
    zodiac = "말"
else:  # remainder == 11
    zodiac = "양"

# 결과 출력
print(f"{year}년도의 태생은 {zodiac}띠입니다.")
```

[문제12] A 학급에 총 10명의 학생이 있다. 이 학생들의 중간고사 점수는 다음과 같다.  
[70, 60, 55, 75, 95, 90, 80, 80, 85, 100]

for문을 사용하여 A 학급의 평균 점수를 구해 보자.

```python
# 점수 리스트
scores = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]

# 리스트 컴프리헨션을 사용해 총점 계산
total = sum([score for score in scores])

# 평균 계산
average = total / len(scores)

# 결과 출력
print("A 학급의 평균 점수는", average, "입니다.")
```

문제13) 
======음식점 메뉴 관리 프로그램======
1. 신규 메뉴 등록
2. 메뉴판 전체 보기
3. 프로그램 종료
> 1
메뉴명: 짜짜
가격: 3000
신규 메뉴 짜짜(이)가 등록되었습니다.
{'짜짜': '3000'}

======음식점 메뉴 관리 프로그램======
1. 신규 메뉴 등록
2. 메뉴판 전체 보기
3. 프로그램 종료
> 2

==========메뉴판==========
짜짜 : 3000원
----------------------------------
1. 수정 | 2. 삭제 | 3. 나가기
> 3

======음식점 메뉴 관리 프로그램======
1. 신규 메뉴 등록
2. 메뉴판 전체 보기
3. 프로그램 종료
> 3
프로그램을 종료합니다.

```python
foods = {}

while True:
    print("\n======음식점 메뉴 관리 프로그램======")
    print("1. 신규 메뉴 등록")
    print("2. 메뉴판 전체 보기")
    print("3. 프로그램 종료")
    num = int(input("> "))

    if num == 1:
        # 1. 메뉴명을 입력받아 해당메뉴가 사전에 키로 존재한다면
        #    "이미 등록된 메뉴입니다"를 출력하세요
        # 2. 사전에 존재하지 않는 메뉴(key)라면 가격을 입력받아
        #    key:value쌍으로 맵핑하여 사전에 저장하세요.
        f_name = input("메뉴명: ")

        if f_name not in foods:
            price = input("가격: ")
            foods[f_name] = price
            print("신규 메뉴 %s(이)가 등록되었습니다." % f_name)
        else:
            print("이미 등록된 메뉴입니다.")

        print(foods)

    elif num == 2:
        # 1. 모든 키를 뽑아내서 리스트로 만드세요.
        # 2. 해당 리스트를 반복문으로 순회하여 메뉴와 가격을 출력하세요.
        #  ex) 출력 예시: "김치찌개 : 7000원"
        menulist = foods.keys()

        print("\n==========메뉴판==========")
        for m in menulist:
            print("%s : %s원" % (m, foods[m]))
        print("----------------------------------")
        print("1. 수정 | 2. 삭제 | 3. 나가기")
        select = int(input("> "))

        if select == 1:
            name = input("가격을 변경할 메뉴의 이름을 입력하세요: ")

            # 사전에 해당 메뉴가 존재하는지 확인
            if name in foods:
                newprice = input("변경할 가격 입력: ")
                # 해당 메뉴이름의 가격 데이터 newprice로 수정
                foods[name] = newprice
                print("%s의 가격이 %s원으로 변경되었습니다." % (name, newprice))
            else :
                print("%s(은)는 등록된 메뉴가 아닙니다." % name)
        elif select == 2:
            name = input("삭제할 메뉴의 이름을 입력하세요: ")

            # 사전에 해당 메뉴가 존재하는지 확인
            if name in foods:
                # 해당 메뉴 데이터 삭제
                del(foods[name])
                print("%s이(가) 삭제되었습니다." % name)
            else:
                print("%s(은)는 등록된 메뉴가 아닙니다." % name)
        elif select == 3:
            continue

    elif num == 3:
        print("프로그램을 종료합니다.")
        break
```
